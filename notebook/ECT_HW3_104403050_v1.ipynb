```python
# IMPORT LIBRARIES
import numpy as np
import pandas as pd
from apyori import apriori
from pandas.core.frame import DataFrame
```

```python
# LOAD THE DATASET
data = pd.read_csv('CreditCardPromotion.csv')
```

```python
# PREPROCESSING
arr = list(data.values)

# APRIORI

association_rules = apriori(arr, min_support = 0.2, min_confidence = 0.9, max_length = 3) 
association_results = list(association_rules)

DataFrame(association_results)
```
