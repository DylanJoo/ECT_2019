```python
import pandas as pd
import numpy as np
```

Q1-a

```python
data = pd.read_csv('fifa19.csv')
```

```python
# Select the mentioned features(a.k.a Name、Age、Wage、Skill Moves、Work Rate、Height、Weight、 Crossing ~ GKReflexes), and drop the other
columns = np.array(data.columns)
# Copy the result from columns label array, and then select the required columns.
true = ['Name','Age','Wage','Skill Moves','Work Rate','Height','Weight',
        'Crossing', 'Finishing', 'HeadingAccuracy',
       'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed',
       'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping',
       'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions',
       'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking',
       'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
datax = data[true]
datax.shape  #41 features
```

Q1-b

```python
datax['Work Rate'].value_counts()
OH_WR = datax['Work Rate'].str.get_dummies(sep='/ ')
# http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.get_dummies.html
datax = pd.concat([datax, OH_WR], 1)
```

```python
# To keep the order(if needed...)
order = np.array(datax.columns)
order = np.delete(order, 4)
order = np.insert(order, 4, ['High', 'Medium', 'Low'])
order = order[:-3]
datax = datax[order]
```

Q1-c

```python
# First look at height, 48 instances wo/
datax['Height'].isnull().sum() #48
datax = datax[~datax['Height'].isnull()] # after removal: (18207-48)
datax['Weight'].isnull().sum() # No any null value
```

Q1-d

```python
# Dealing with numpy array
w = np.array(datax['Wage'])
w = w.astype('str')
w = np.char.strip(w, '€K')
# NTD:EU = 35:1 , K means a thou
w = w.astype('int32')
w = w*35*1000
# Replace by new wage array
datax['Wage'] = w
```

Q1-e

```python
# Dealing with series this time ;)
datax['Height'] = datax['Height'].astype('str')
# Expand into two array, represented the feets and inches
h = datax['Height'].str.split("'", expand = True)
h = h.astype('float64')
h = h[0]*30.48 + h[1]*2.54
# Replace by new height
datax['Height'] = h

# Rename it
datax = datax.rename(columns={'Height' : 'Height(cm)'})
```

Q1-f

```python
# Dealing with series this time ;)
datax['Weight'] = datax['Weight'].astype('str')
# Expand into two array, represented the feets and inches
h = datax['Weight'].str.strip("lbs")
h = h.astype('float64')
h = h*0.4536
# Replace by new height array
datax['Weight'] = h

# Rename it
datax =datax.rename(columns = {'Weight': 'Weight(kg)'})
```

Q1-g

```python
datax.to_csv('104403050_data_new.csv')
```
